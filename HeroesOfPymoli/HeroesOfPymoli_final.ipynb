## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# establish end point for bin
max_age = purchase_data_df["Age"].max()
# Creates bins
bins= [0,9,14,19,24,29,34,39,max_age]
print(bins)
```

```python
# Create labels for bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
print(group_names)
```

```python
# Categorize the existing players
age_demo_df = purchase_data_df
age_demo_df["Age Range"] = pd.cut(age_demo_df["Age"], bins, labels = group_names)
age_demo_df
```

```python
# Count numbers by age group
age_demo_df = age_demo_df.groupby('Age Range')
total_age_group = age_demo_df["SN"].nunique()
total_age_group
```

```python
# Calcualte percentages by age group (use total players counted from player_count)
percentage_of_players = (total_age_group/player_count) * 100
percentage_of_players
```

```python
# Create Dataframe
age_demo_df = pd.DataFrame({"Total Count": total_age_group, 
                            "Percentage of Players": percentage_of_players})
# Format DataFrame
age_demo_df["Percentage of Players"] = age_demo_df["Percentage of Players"].map("{:.2f}%".format)
# Remove the index name "Age Range"
age_demo_df.index.name = None
age_demo_df
```

```python
# Bin the purchase_data data frame by age
purchase_analysis_df = purchase_data_df
purchase_analysis_df["Age Group"] = pd.cut(purchase_analysis_df["Age"], bins, labels = group_names)
purchase_analysis_df
```

```python
# Calcualte Average Purchase Price 
avg_purchase_price2 = purchase_analysis_df["Price"].mean()
avg_purchase_price2
```

```python
#  Calculate Avg Total Purchase per Person by age group (per total_age_group)
avg_cost_per_person2 = total_purchase_value2 / total_age_group
avg_cost_per_person2
```

```python
# Create DataFrame
purchase_analysis_df = pd.DataFrame({"Purchase Count": purchase_count2, 
                      "Average Purchase Price": avg_purchase_price2,
                      "Total Purchase Value": total_purchase_value2,
                      "Avg Total Purchase per Person": avg_cost_per_person2})
# Optional: Format DataFrame
purchase_analysis_df["Average Purchase Price"] = purchase_analysis_df["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis_df["Total Purchase Value"] = purchase_analysis_df["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis_df["Avg Total Purchase per Person"] = purchase_analysis_df["Avg Total Purchase per Person"].map("${:.2f}".format)
purchase_analysis_df.rename = index= {"Age Range": "Age Group"}
purchase_analysis_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# revisualize dataset
purchase_data_df.head()
```

```python
# extract item id, item name, and price from original dataframe
pop_items_df = purchase_data_df[["Item ID", "Item Name", "Price"]]
pop_items_df.head()
```

```python
# create groupby
pop_items_df = pop_items_df.groupby(["Item ID", "Item Name"])
pop_items_df.head()
```

```python
# Create Purchase Count
purchase_count_col = pop_items_df["Item ID"].count()
purchase_count_col
```

```python
# Create Purchase Value
purchase_value = pop_items_df["Price"].sum()
purchase_value
```

```python
# Item Price (total purchase value/ purchase count)
item_price = purchase_value/ purchase_count_col
item_price
```

```python
# Create DF
pop_items_df = pd.DataFrame({"Purchase Count": purchase_count_col, 
                             "Item Price": item_price,
                             "Total Purchase Value": purchase_value})
# display
pop_items_df.head()
```

```python
# create sorted dataframe
pop_items_df_sort = pd.DataFrame(pop_items_df.copy())
pop_items_df_sort.head()
# Optional: format
pop_items_df_sort["Item Price"] = pop_items_df_sort["Item Price"].map("${:.2f}".format)
pop_items_df_sort["Total Purchase Value"] = pop_items_df_sort["Total Purchase Value"].map("${:.2f}".format)
# sort in descending order
pop_items_df_sort.sort_values("Purchase Count", ascending= False).head()
```
